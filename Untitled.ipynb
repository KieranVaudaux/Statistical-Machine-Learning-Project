```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

colname = list(cdf.columns)
color = list(['green','k','red','blue'])

f = lambda x: 300 + 5.5*x
```

```python
fig = plt.figure(figsize=(8,8))
cdf = FICO["cdf"]
i = 0
for name in colname:
    y = cdf[name]
    plt.plot(f(cdf.index),y,color = color[i],label=name)
    fig.legend(loc = (0.1,0.79))
    i += 1
```

```python
fig = plt.figure(figsize=(8,8))
pdf = FICO["pdf"]
i = 0
for name in colname:
    y = pdf[name]
    plt.plot(f(pdf.index),y,color = color[i],label = name)
    fig.legend(loc = (0.75,0.64))
    i += 1
```

```python
totals = FICO["totals"]
fraction_score_group = pd.DataFrame(index=pdf.index,columns=colname)
for name in colname:
    fraction_score_group[name]=totals[name]*pdf[name]
```

```python
perf = FICO["performance"]

fig = plt.figure(figsize=(10,8))
i = 0
for name in colname:
    y = perf[name]
    plt.plot(f(perf.index),y,color = color[i],label = name)
    fig.legend(loc = (0.75,0.64))
    i += 1
```

```python
# False Positive Rate by score as threshold per group
fpr = FICO["fpr"]

fig = plt.figure(figsize=(10,8))
i = 0
for name in colname:
    y = fpr[name]
    plt.plot(f(fpr.index),y,color = color[i],label = name)
    fig.legend(loc = (0.75,0.64))
    i += 1
```

```python
# True Positive Rate by score as threshold per group
tpr = FICO["tpr"]

fig = plt.figure(figsize=(10,8))
i = 0
for name in colname:
    y = tpr[name]
    plt.plot(f(tpr.index),y,color = color[i],label = name)
    fig.legend(loc = (0.75,0.64))
    i += 1
```
