```python
cdf = FICO["cdf"]
```

```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

colname = list(cdf.columns)
color = list(['green','k','red','blue'])
i = 0

fig = plt.figure(figsize=(8,8))
x = 300 + 5.5*np.array(cdf.index)

for name in colname:
    y = cdf[name]
    plt.plot(x,y,color = color[i])
    i += 1
```
